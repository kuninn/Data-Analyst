CLEANING DATA

```python
#Membersihkan data customers_df
#Menghilangkan Duplikat data
customers_df.drop_duplicates(inplace=True)
```

```python
#Memeriksa apakah masih ada duplikat data pada customers_df
print("Jumlah Duplikat data: ", customers_df.duplicated().sum())
```

```python
#Menangani missing value
#Melihat data yang mengandung missing value

customers_df[customers_df.gender.isna()]
```

```python
#Menghitung value yang dominan
customers_df.gender.value_counts()

#pada kolom berikut, nilai yang paling dominan adalah "Prefer not ot say"
#kita akan menggunakan method fillna()
```

```python
#Method fillna() metode imputasi
customers_df.fillna(value="Prefer not to say", inplace=True)
```

```python
#Untuk memastikan proses berjalan dengan benar
#dan tidak ada missing value
customers_df.isna().sum()
```

```python
customers_df[customers_df.gender.isna()]
```

```python
#Menangani value yang tidak akurat
customers_df[customers_df.age == customers_df.age.max()]
```

```python
#Proses yang dilakukan menggunakan replace
customers_df.age.replace(customers_df.age.max(), 80, inplace=True)
```

```python
customers_df[customers_df.age == customers_df.age.max()]
```

```python
#untuk memastikan tidak ada inacurate value
customers_df.describe()
```

```python
#Membersihkan data orders_df
#yang memiliki kesalahan tipe data pada date_time

datetime_coloumns =["order_date", "delivery_date"]

for coloumn in datetime_coloumns:
  orders_df[coloumn] = pd.to_datetime(orders_df[coloumn])
```
